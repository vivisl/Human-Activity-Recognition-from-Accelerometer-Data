```python
%matplotlib inline
import pathlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.lib.stride_tricks import as_strided
from mpl_toolkits.mplot3d import Axes3D
from time import time
from scipy.stats import randint as sp_randint
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier


class Recognition:
    
    @staticmethod
    def label(data, l):
        cut = data[data['label']==l]
        skip = int(cut.shape[0] * 0.1) 
        cut = cut.iloc[skip:-skip, :]
        return cut
    
    @staticmethod
    def filterPass(data, l):
        xyzm = Recognition.label(data, l).drop('label', axis=1)
        b = xyzm.values
        freq = np.fft.fft(xyzm)
        lowPassMask = abs(freq) < 1     
        lowPassFy = freq.copy()
        lowPassFy[~lowPassMask] = 0    
        dc = np.fft.ifft(lowPassFy).real
        highPassFy = freq.copy()
        highPassFy[lowPassMask] = 0
        ac = np.fft.ifft(highPassFy).real 
        bdcac = np.hstack((b,dc,ac))     
        return bdcac    

    @staticmethod
    def figurePlot(data, l):
        bdcac = Recognition.filterPass(data, l)
        for i in range(0,4):
            plt.subplot(221+i)
            plt.plot(bdcac[:,i], label='b')
            plt.plot(bdcac[:,i+4], label='dc')
            plt.plot(bdcac[:,i+8], label='ac')
            if i == 0:
                name = 'x'
            elif i == 1:
                name = 'y'
            elif i == 2:
                name = 'z'
            else:
                name = 'm'
            plt.title(str(l)+'_'+name) 
            plt.legend(loc='upper right')
            plt.show() 
    
    @staticmethod
    def windowed_view(arr, window, overlap):
        arr = np.asarray(arr)
        window_step = window - overlap
        new_shape = arr.shape[:-1] + ((arr.shape[-1] - overlap) // window_step, window)
        new_strides = (arr.strides[:-1] + (window_step * arr.strides[-1],) + arr.strides[-1:])
        return as_strided(arr, shape=new_shape, strides=new_strides)
    
    @staticmethod
    def feature_generate(data, l):
        meanlst = []
        stdlst = []
        minmaxlst = []
        rmslst = []
        bdcac = Recognition.filterPass(data, l)
        for i in range(0,12): 
            window = Recognition.windowed_view(bdcac[:,i], 52, 26)
            # mean
            temp = np.mean(window, axis=1)
            meanlst.append(temp)
            # std
            temp = np.std(window, axis=1)
            stdlst.append(temp)
            # minmax
            temp = np.amax(window, axis=1) - np.amin(window, axis=1)
            minmaxlst.append(temp)
            # rms        
            rms_win1 = Recognition.windowed_view(bdcac[:,i], 10, 9)
            series1 = np.sum(rms_win1, axis=1)
            rms_win2 = Recognition.windowed_view(series1, 52, 26)
            series2 = np.sum(rms_win2**2, axis=1)
            series3 = np.sqrt(series2/len(series2))
            rmslst.append(series3)
        # convert list to array and transpose it to have 12 dimonsions which is {b,dc,ac}*{x,y,z,m}    
        rms = np.asarray(rmslst).T
        # somtimes rms has different dimonsions from mean,std and minmas 
        # so convert mean,std and minmas to have the same dimonsions as rms
        mean = np.asarray(meanlst).T[:rms.shape[0],:]        
        std = np.asarray(stdlst).T[:rms.shape[0],:]
        minmax = np.asarray(minmaxlst).T[:rms.shape[0],:]
        features = np.hstack((mean,std,minmax,rms))    
        return features
    
    @staticmethod
    def df(data, l):        
        features = Recognition.feature_generate(data, l)
        df = pd.DataFrame()
        
        df['Mean_xb'] = features[:,0]
        df['STD_xb'] = features[:,12]
        df['Minmax_xb'] = features[:,24]
        df['RMS_xb'] = features[:,36]  
        
        df['Mean_yb'] = features[:,1]
        df['STD_yb'] = features[:,13]
        df['Minmax_yb'] = features[:,25]
        df['RMS_yb'] = features[:,37]  
        
        df['Mean_zb'] = features[:,2]
        df['STD_zb'] = features[:,14]
        df['Minmax_zb'] = features[:,26]
        df['RMS_zb'] = features[:,38] 
        
        df['Mean_mb'] = features[:,3]
        df['STD_mb'] = features[:,15]
        df['Minmax_mb'] = features[:,27]
        df['RMS_mb'] = features[:,39]    
    
        df['Mean_xdc'] = features[:,4]
        df['STD_xdc'] = features[:,16]
        df['Minmax_xdc'] = features[:,28]
        df['RMS_xdc'] = features[:,40] 
        
        df['Mean_ydc'] = features[:,5]
        df['STD_ydc'] = features[:,17]
        df['Minmax_ydc'] = features[:,29]
        df['RMS_ydc'] = features[:,41]  
        
        df['Mean_zdc'] = features[:,6]
        df['STD_zdc'] = features[:,18]
        df['Minmax_zdc'] = features[:,30]
        df['RMS_zdc'] = features[:,42]  
        
        df['Mean_mdc'] = features[:,7]
        df['STD_mdc'] = features[:,19]
        df['Minmax_mdc'] = features[:,31]
        df['RMS_mdc'] = features[:,43]   
        
        df['Mean_xac'] = features[:,8]
        df['STD_xac'] = features[:,20]
        df['Minmax_xac'] = features[:,32]
        df['RMS_xac'] = features[:,44]  
        
        df['Mean_yac'] = features[:,9]
        df['STD_yac'] = features[:,21]
        df['Minmax_yac'] = features[:,33]
        df['RMS_yac'] = features[:,45]  
        
        df['Mean_zac'] = features[:,10]
        df['STD_zac'] = features[:,22]
        df['Minmax_zac'] = features[:,34]
        df['RMS_zac'] = features[:,46]  
        
        df['Mean_mac'] = features[:,11]
        df['STD_mac'] = features[:,23]
        df['Minmax_mac'] = features[:,35]
        df['RMS_mac'] = features[:,47]   
       
        df['label'] = l
        return df
    

# put 1.csv to 15.csv into same folder by yourself   
path1 = './data1'
path2 = './data2'
inpath = pathlib.Path(path1)
file_list = sorted(inpath.glob('*.csv'))
features_list = []
all_features = pd.DataFrame()
for f in file_list:            
    print('read', f)
    temp = pd.read_csv(f, usecols=[1,2,3,4], header=None, names = ['x','y','z','label'])          
    temp['m'] = np.sqrt(temp['x']**2 + temp['y']**2 + temp['z']**2) 
    for l in range (1,8):
        Recognition.figurePlot(temp, l)
        features = Recognition.df(temp, l)
        features_list.append(features) 
all_features = all_features.append(features_list)
all_features.to_csv('all_features.csv', index=False)
print('{} feature files are concatenated and saved as all_features.csv: done \n'.format(len(file_list)))
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                    results['mean_test_score'][candidate],
                    results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")            
            
x = all_features.drop('label', axis=1)
y = all_features['label']
clf = RandomForestClassifier(n_estimators=20)
            
# specify parameters and distributions to sample from and run randomized search
param_dist = {"max_depth": [10, None],
              "max_features": sp_randint(2, 30), 
              "min_samples_split": sp_randint(2, 30),
              "min_samples_leaf": sp_randint(2, 30),
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}
n_iter_Random= 30
random_search = RandomizedSearchCV(clf, param_distributions=param_dist, n_iter=n_iter_Random)
start = time()
random_search.fit(x, y)
print("RandomizedSearchCV took %.2f seconds for %d candidates parameter settings." % (time()-start, n_iter_Random))
report(random_search.cv_results_)

# use a full grid over all parameters and run grid search
param_grid = {"max_depth": [10, None],
              "max_features": [2, 15, 30],
              "min_samples_split": [2, 15, 30],
              "min_samples_leaf": [2, 15, 30],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}
grid_search = GridSearchCV(clf, param_grid=param_grid)
start = time()
grid_search.fit(x, y)
n_iter_grid = len(grid_search.cv_results_['params'])
print("GridSearchCV took %.2f seconds for %d candidate parameter settings." % (time()-start, n_iter_grid))
report(grid_search.cv_results_)
```
